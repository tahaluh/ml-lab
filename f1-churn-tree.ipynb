```python
features = [
    'qtRaces', 'qtTeams',
    'avgPosition',
    'avgPositionLift',
    'pctPolePositition',
    'avgPositionRaceLast10',
    'avgPositionLiftRaceLast10',
    'pctPolePosititionRaceLast10',
    'avgPositionLiftRaceLast40',
    'flChurn'
]

#clean_train = df_train[features]
clean_train = df_train.copy().select_dtypes(include=["number"])

display(clean_train.head())
```

```python
numeric_df = clean_train.select_dtypes(include=["number"])

correlation_matrix = numeric_df.corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 12))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=False, fmt=".2f")
plt.title("Matriz de Correlao")
plt.show()
```

```python
np.fill_diagonal(correlation_matrix.values, 0)

high_corr = correlation_matrix[correlation_matrix > 0.8]

high_corr_pairs = high_corr.unstack().dropna().sort_values(ascending=False).drop_duplicates()
print(high_corr_pairs)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
y=clean_train["flChurn"]
X=clean_train.drop(["flChurn"], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=22)


display(X.head())
print(y.head())
```

```python
model = RandomForestClassifier(random_state=42, class_weight='balanced', max_depth=10, n_estimators=100)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
best_acc = 0
best_tree_idx = 0

for i, tree in enumerate(model.estimators_):
    pred = tree.predict(X_test)
    acc = accuracy_score(y_test, pred)
    if acc > best_acc:
        best_acc = acc
        best_tree_idx = i

print(f"Melhor rvore: #{best_tree_idx}, Acurcia: {best_acc}")
```

```python
model = DecisionTreeClassifier(random_state=42, class_weight='balanced', max_depth=10)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'max_depth': [4, 6, 8],
    'n_estimators': [50, 100],
    'min_samples_leaf': [1, 5, 10]
}

grid = GridSearchCV(
    RandomForestClassifier(class_weight='balanced', random_state=42),
    param_grid,
    scoring='precision',  #  foco em precision
    cv=5
)
grid.fit(X_train, y_train)
```
